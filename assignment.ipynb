```python
import pandas as pd
import numpy as np
```

```python
input = pd.read_excel("Assignment_Timecard.xlsx")

df_cleaned = input.dropna(subset=['Position ID','Position Status','Time',	'Time Out'	,'Timecard Hours (as Time)'	,'Pay Cycle Start Date',	'Pay Cycle End Date',	'Employee Name',	'File Number'], how='any')
df_cleaned=df_cleaned.dropna(axis=1)
df_cleaned
```

```python
# a) who has worked for 7 consecutive days.


unique_names = df_cleaned['Employee Name'].unique()
df = df_cleaned.sort_values(by='Time')

# Identify consecutive dates for each person
consecutive_dates = df.groupby('Employee Name')['Time'].diff().ne(1).cumsum()

# Iterate through unique names
for name in df['Employee Name'].unique():
    # Get the corresponding dates for the current name
    name_dates = df[df['Employee Name'] == name]['Time']

    # Check if the person has worked for 7 consecutive days
    consecutive_counts = name_dates.diff().ne(1).cumsum()
    seven_consecutive_days = consecutive_counts.value_counts().get(1, 0) >= 7

    # Print or use the results as needed
    if(seven_consecutive_days==0):
        print(f"{name} has worked for 7 consecutive days: {seven_consecutive_days}")

```

```python
# b) who have less than 10 hours of time between shifts but greater than 1 hour

df = df_cleaned.sort_values(by=['Employee Name', 'Time'])

# Identify consecutive dates for each person
consecutive_dates = df.groupby('Employee Name')['Time'].diff().ne(1).cumsum()

# Iterate through unique names
for name in df['Employee Name'].unique():
    # Get the corresponding dates for the current name
    name_dates = df[df['Employee Name'] == name]['Time']

    # Calculate the time difference between shifts
    time_diff = name_dates.diff()

    # Filter shifts with less than 10 hours but greater than 1 hour between them
    selected_shifts = name_dates[(time_diff < pd.Timedelta(hours=10)) & (time_diff > pd.Timedelta(hours=1))]

    # Print the result
    print(f"{name} has shifts with less than 10 hours but greater than 1 hour between them:")
    #print(selected_shifts)
```

```python
# c) Who has worked for more than 14 hours in a single shift

for name in df['Employee Name'].unique():
    # Get the corresponding time in and time out for the current name
    name_data = df[df['Employee Name'] == name][['Time', 'Time Out']]

    # Calculate the duration of each shift
    name_data['shift_duration'] = pd.to_datetime(name_data['Time Out']) - pd.to_datetime(name_data['Time'])

    # Filter shifts with more than 14 hours of duration
    long_shifts = name_data[name_data['shift_duration'] > pd.Timedelta(hours=14)]

    # Print the result
    if not long_shifts.empty:
        print(f"{name} has worked for more than 14 hours in a single shift:")
        print(long_shifts[['Time', 'Time Out', 'shift_duration']])
```
